```python
#removing the 228 row from the dataframe
df_gdp_final = df_gdp.head(df_gdp.shape[0] -1)
df_gdp_final
```

```python
y_2023 = df_gdp_final[['Country', '2023']].sort_values(by ='2023',ascending = True).tail(7)
```

```python
fig, ax = plt.subplots(figsize = (11,6))

##creating the bar plot
plt.barh(y_2023["Country"],y_2023["2023"], color = 'darkorchid')
plt.title("Top countries with highest GDP in 2023", color = 'darkorchid', fontweight = 'bold', loc ='left', fontsize = 19)
plt.ylabel("Country", fontsize = 14 )
plt.xlabel("per capita GDP (US$)", fontsize = 14)
plt.legend([2023], fontsize = 14)

ax.set_facecolor('white')
ax.set_facecolor('white')
ax.spines["bottom"].set_color("black")      # x axis line
ax.spines["top"].set_color("black")
ax.spines["left"].set_color("black")
ax.spines["right"].set_color("black")

#add watermark on figure
fig.text(0.87, 0.2, '@shrutis', fontsize = 10, color ='grey', ha ='right', va ='bottom')

plt.show()
fig.savefig("gdp_2023.png")
#files.download("avc.png")
```

```python
z_2022 = df_gdp_final[['Country', '2022']].sort_values(by ='2022',ascending = True).tail(7)
```

```python
fig1, ax1 = plt.subplots(figsize = (11,6))

#creating the bar plot
plt.barh(z_2022["Country"],z_2022["2022"], color = 'rebeccapurple')
plt.title("Top countries with highest GDP in 2022", color = 'rebeccapurple', fontweight = 'bold', loc ='left', fontsize = 19)
plt.ylabel("Country", fontsize = 14 )
plt.xlabel("per capita GDP (US$)", fontsize = 14)
plt.legend([2022], fontsize = 14)

ax1.set_facecolor('white')
ax1.spines["bottom"].set_color("black")      # x axis line
ax1.spines["top"].set_color("black")
ax1.spines["left"].set_color("black")
ax1.spines["right"].set_color("black")

#add watermark on figure
fig1.text(0.87, 0.2, '@shrutis', fontsize = 10, color ='grey', ha ='right', va ='bottom')

plt.show()
fig1.savefig("gdp_2022.png")
#files.download("abc.png")
```
